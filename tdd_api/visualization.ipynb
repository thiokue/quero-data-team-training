```python
import polars as pl
import matplotlib.pyplot as plt
import os
```

```python
data_folder = '/home/thiagouehara/Documentos/ex-lucas/quero-data-team-training/tdd_api/data/gold'
files = [os.path.join(data_folder, f) for f in os.listdir(data_folder) if f.endswith('.parquet')]
print(files)
```

```python
data_frames = [pl.read_parquet(file) for file in files]
print(data_frames)
```

```python
if len(data_frames) > 1:
    data = pl.concat(data_frames, ignore_index=True)
else:
    data = data_frames[0]

data = data.sort('date')

plt.figure(figsize=(10, 5))
plt.plot(data['date'], data['price'], label='Gold Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Historical Gold Prices')
plt.legend()
plt.show()
```
